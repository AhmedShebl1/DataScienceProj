```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import cross_val_score, cross_val_predict
from sklearn.linear_model import LinearRegression 
from sklearn import metrics
import warnings

warnings.simplefilter(action='ignore', category=FutureWarning)

#load wrngled dataset
df = pd.read_csv('student-por.csv')
df
```

```python
#Perform one hot encode method for nominal data and label encoding for ordinal data
numerical_fields = ['age', 'absences', 'G1', 'G2', 'G3', 'traveltime', 'studytime', 'freetime']
categorical_ordinal_fields = ['Medu', 'Fedu', 'failures', 'famrel', 'goout', 'Dalc', 'Walc', 'health']
categorical_nominal_fields = ['school', 'sex', 'address', 'famsize', 'Pstatus', 'Mjob', 'Fjob', 'reason', 'guardian', 'schoolsup', 'famsup', 'paid', 'activities', 'nursery', 'higher', 'internet', 'romantic']
nominal_non_numeric = []
label_encoder = LabelEncoder()

for col in df.columns:
    if df[col].dtype == 'object':
        if (col in categorical_ordinal_fields):
            df[col] = label_encoder.fit_transform(df[col])
        elif (col in categorical_nominal_fields):
            nominal_non_numeric.append(col)

df = pd.get_dummies(df, columns=nominal_non_numeric)


df
```

```python
#get the correlations between all the fields and display correlations matrix
correlation = df.corr()
correlation
```

```python
#choose significant features based on correlation
G3_corr = correlation['G3']
significant_features = G3_corr[(abs(G3_corr) > 0.15) & (G3_corr != 1)].index.tolist()
significant_features 
```

```python
#form a matrix from chosen signifacnt features

X = df[significant_features]
y = df['G3']

# Normalize numerical fields

numerical_significant = []

for field in numerical_fields:
    if (field in X):
        numerical_significant.append(field)
        

if (len(numerical_significant) != 0):
    scaler = StandardScaler()
    X.loc[:, numerical_significant] = scaler.fit_transform(X[numerical_significant])

#create a linear model

model = LinearRegression()

#perform K fold cross validation method for the best model accuracy

cv_predictions = cross_val_predict(model, X, y, cv=5)
cv_scores = cross_val_score(model, X, y, cv=5, scoring='neg_mean_squared_error')
cv_mse = -np.mean(cv_scores)
cv_r2 = cv_scores = cross_val_score(model, X, y, cv=5, scoring='r2')

model.fit(X, y)
feature_coefficients = model.coef_
intercept = model.intercept_

#show results

print(f"Cross-Validation Mean Squared Error: {cv_mse}")
print(f"Cross-Validation Root Mean Squared Error: {cv_mse**(0.5)}")
print(f"Cross-Validated R Score: {np.mean(cv_r2)}")
print('-----------------------------------------------------------')

for i, coefficient in enumerate(feature_coefficients):
    print(f"Feature : {X.columns[i]} , Coefficient : {coefficient}")


print("Intercept:", intercept)
```
