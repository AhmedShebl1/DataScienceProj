```python
from scipy.stats import ttest_rel

undergrad_scores = df['etest_p']
mba_scores = df['mba_p']

t_statistic, p_value = ttest_rel(undergrad_scores, mba_scores)

alpha = 0.05

print("Paired t-test results:")

print("p-value:", p_value)

if p_value < alpha:
    print("Reject the null hypothesis. There is a significant difference in mean percentage scores between undergraduate degree and MBA.")
else:
    print("Fail to reject the null hypothesis. There is no significant difference in mean percentage scores between undergraduate degree and MBA.")
```
