```python
# Calculate the outliers
Q1 = df['Height(CM)'].quantile(0.25)
Q3 = df['Height(CM)'].quantile(0.75)

IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

print(upper_bound)
print(lower_bound)

outliers = df[(df['Height(CM)'] < lower_bound) | (df['Height(CM)'] > upper_bound)]
outliers
```

```python
plt.scatter(df[(df['Height(CM)'] > 125) & (df['Height(CM)'] < 197)]['Height(CM)'],
            df[(df['Height(CM)'] > 125) & (df['Height(CM)'] < 197)]['college mark'])
```

```python
plt.scatter(df['Height(CM)'],df['college mark'])
plt.title('Scatter Plot of Height and college marks')
plt.xlabel('Height(CM)')
plt.ylabel('colllege mark')
```
