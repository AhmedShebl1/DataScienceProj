```python
import numpy as np
import pandas as  pd
import matplotlib.pyplot as plt
import scipy.stats as stats

df = pd.read_excel('C:/Users/www/Downloads/saa.xlsx')
df
```

```python
df.columns = ['','Certification Course','Gender','Department','Height(CM)','Weight(KG)','10th Mark','12th Mark','college mark','hobbies','daily studing time','prefer to study in','salary expectation','Do you like your degree?','willingness to pursue a career based on their degree ' ,'social medai & video','Travelling Time ','Stress Level ','Financial Status','part-time job']
```

```python
df
```

```python
sampled_df = df['Weight(KG)'].sample(frac=0.25,random_state = 5)
sample_mean = np.mean(sampled_df)
population_mean = np.mean(df['Weight(KG)'])
print(population_mean)
print(sample_mean)
```

```python
sample_size = len(sampled_df)
print(sample_size)
Square_error = (np.std(df['Weight(KG)'],ddof=1))/(np.sqrt(sample_size))
print(Square_error)
```

```python
Z_score = (sample_mean-population_mean)/Square_error
p_value = 2 * stats.norm.cdf(-abs(Z_score))

print("The p value for this hypothesis test is: {:20.19f}".format(p_value))
```

```python
alpha = 0.05
if p_value <= alpha/2:
    print("The null hypothesis is rejected as there is evidence to support the alternative hypothesis in the left tail.")
elif p_value >= 1 - alpha/2:
    print("The null hypothesis is rejected as there is evidence to support the alternative hypothesis in the right tail.")
else:
    print("The null hypothesis cannot be rejected as there is not enough evidence to support the alternative hypothesis.")
```
