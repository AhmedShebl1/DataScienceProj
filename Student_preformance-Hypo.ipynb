```python
import numpy as np
import pandas as  pd
import matplotlib.pyplot as plt
import scipy.stats as stats

df = pd.read_csv('student-por.csv')
df
```

```python
sampled_df = df['absences'].sample(frac=0.25,random_state = 5)
sample_mean = np.mean(sampled_df)
population_mean = np.mean(df['absences'])
print("Population mean = ",population_mean)
print("Sample mean = ",sample_mean)

sample_size = len(sampled_df)
print("sample size = ",sample_size)
Square_error = (np.std(df['absences'],ddof=1))/(np.sqrt(sample_size))
print("Square error = ",Square_error)

Z_score = (sample_mean-population_mean)/Square_error
p_value = 2 * stats.norm.cdf(-abs(Z_score))
print("")
print("The Z score for this hypothesis test is: {:20.19f}".format(Z_score))
print("The p value for this hypothesis test is: {:20.19f}".format(p_value))

print("")
alpha = 0.05
if p_value <= alpha/2:
    print("The null hypothesis is rejected as there is evidence to support the alternative hypothesis in the left tail.")
elif p_value >= 1 - alpha/2:
    print("The null hypothesis is rejected as there is evidence to support the alternative hypothesis in the right tail.")
else:
    print("The null hypothesis cannot be rejected as there is not enough evidence to support the alternative hypothesis.")
```
